```python
# При необходимости изменит епуть к данным
data = pd.read_csv("/content/adult.data.csv")
data.head()
```

**1. Сколько мужчин и женщин (признак *sex*) представлено в этом наборе данных?**

**4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год? **

```python
# более 50к в год
high_income = data[data['salary'] == '>50K']
mean_high = high_income['age'].mean()
std_high = high_income['age'].std()
print("Средний возраст (>50K):", mean_high)
print("Стандартное отклонение (>50K):", std_high)
# менее 50к в год

low_income = data[data['salary'] == '<=50K']
mean_low = low_income['age'].mean()
std_low = low_income['age'].std()
print("Средний возраст (<=50K):", mean_low)
print("Стандартное отклонение (<=50K):", std_low)
```

**6. Правда ли, что люди, которые получают больше 50k, имеют как минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

```python
higher_ed = [
    'Bachelors',
    'Prof-school',
    'Assoc-acdm',
    'Assoc-voc',
    'Masters',
    'Doctorate'
]

# Фильтруем людей с доходом >50K
high_income = data[data['salary'] == '>50K']

# Проверяем, есть ли среди них люди без высшего образования
violations = high_income[~high_income['education'].isin(higher_ed)]

print("Количество людей >50K без высшего образования:", len(violations))
print("Есть ли такие люди:", len(violations) > 0)
```

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

```python
age_stats = data.groupby(['race', 'sex'])['age'].describe()
print(age_stats)

max_age = data[(data['race'] == 'Amer-Indian-Eskimo') & (data['sex'] == 'Male')]['age'].max()
print("\nМаксимальный возраст мужчин Amer-Indian-Eskimo:", max_age)
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
males = data[data['sex'] == 'Male']

married_males = males[males['marital-status'].str.startswith('Married')]

single_males = males[~males['marital-status'].str.startswith('Married')]

rich_married = (married_males['salary'] == '>50K').mean()

rich_single = (single_males['salary'] == '>50K').mean()

print("Доля богатых среди женатых мужчин:", rich_married)
print("Доля богатых среди холостых мужчин:", rich_single)
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python

max_hours = data['hours-per-week'].max()
print("Максимальное число часов в неделю:", max_hours)

workers_max_hours = data[data['hours-per-week'] == max_hours]
count_workers = len(workers_max_hours)
print("Количество людей, работающих столько часов:", count_workers)

rich_share = (workers_max_hours['salary'] == '>50K').mean()
print("Доля зарабатывающих >50K среди них:", rich_share)
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
data['native-country'] = data['native-country'].replace('?', 'Unknown')

result = data.groupby(['native-country', 'salary'])['hours-per-week'].mean().reset_index()

print(result.rename(columns={'hours-per-week': 'mean_hours_per_week'}))
```
